```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

# Create a plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, color=(7/255, 82/255, 154/255), marker='o', markersize=14, linestyle = "None")  # Using RGB color and 'o' marker

# Customize plot
plt.title("Circle Markers with RGB Color (7/82/154)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```
